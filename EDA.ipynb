```python
# Так как количественные показатели не имеют тип данных int или float гистограмму можно только по признаку - Высота потолков, м построить 
data['Высота потолков, м'].hist(bins=80,  range = (10,500));
```

```python
# Построем круглые диаграммы по некоторым признакам
fig, axes=plt.subplots(1, 2, figsize=(15, 5))
data['Парковка'].value_counts().plot(kind = 'pie', autopct = '%1.0f%%', title = 'Парковка', ax=axes[0])
data['Ремонт'].value_counts().plot(kind = 'pie', autopct = '%1.0f%%', title = 'Ремонт', ax=axes[1])
fig, axes=plt.subplots(1, 2, figsize=(15, 5))
data['Балкон'].value_counts().plot(kind = 'pie', autopct = '%1.0f%%', title = 'Балкон', ax=axes[0])
data['Можно с детьми/животными'].value_counts().plot(kind = 'pie', autopct = '%1.0f%%', title = 'Можно с детьми/животными ', ax=axes[1])
plt.show();
```
